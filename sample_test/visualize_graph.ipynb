```python
import numpy as np
import matplotlib.pyplot as plt

x1 = np.linspace(-5, 5, 101)
y1 = np.sin(x1)

figure, axis = plt.subplots()
axis.set_title("Sin")
axis.set_xlabel("rad")
axis.plot(x1, y1, label="Sin")

# Set legend
handles, labels = axis.get_legend_handles_labels()
axis.legend(handles, labels)

plt.show()
```
